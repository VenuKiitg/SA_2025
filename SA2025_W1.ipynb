### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import seaborn as sns
import matplotlib.pyplot as plt

df7 = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
# df7 contains average mpg for each model year and origin
plt.figure(figsize=(10, 6))
sns.lineplot(data=df7, x='model_year', y='mpg', hue='origin',marker='s', style='origin', markersize=8, dashes=False)

plt.title('Average mpg over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average mpg')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
hue_order = ['japan', 'europe', 'usa']

df_plot = df.reset_index(drop=False)

plt.figure(figsize=(8, 6))
scatter = sns.scatterplot(
    data=df_plot,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(30, 180),
    hue_order=hue_order,
    alpha=0.6,
    palette='Set2'
)

plt.title('Horsepower vs Weight by Origin (size = mpg)', fontsize=14)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight', fontsize=12)
plt.legend(title='Origin', loc='upper right')
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
df7 = df.groupby('name')['model_year'].nunique().reset_index(name='number_of_times')

df8 = df7[df7['number_of_times'] > 1]

print('1st part of task')
print(df8.head())
print()
# Taking only those cars that have appeared in more than one model year in df = df9
df9 = df[df['name'].isin(df8['name'])]

df10 = df9.groupby('name').agg(
    appearances=('model_year', 'count'),
    mpg_std=('mpg', 'std'),
    mpg_avg=('mpg', 'mean')
).reset_index()
#Computing standard deviation of mpg(2nd part of the task) and average mpg(4th part of the task) for each name

df11 = df10[df10['mpg_std'] < 1.0]
#cars with atleast 2 appearances and mpg std < 1.0(3rd part of the task)

lowest_variation = df11['mpg_std'].min()

df12 = df11[df11['mpg_std'] == lowest_variation]
print('3rd part of task')
print(df12.head())
print()

# 4th part of the task
df13= df10[['name', 'appearances', 'mpg_avg']].copy()
print('4th part of task')
print(df13.head())
print()

#Bonus part of the task
# Sorting the DataFrame by 'appearances' in descending order then 'mpg_avg' in descending order
df14 = df13.sort_values(by=['appearances', 'mpg_avg'], ascending=[False, False])
print('Bonus part of task')
print(df14.head())  
```
